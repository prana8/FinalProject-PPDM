```python
import pandas as pd
import re
import nltk
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import math

nltk.download('punkt')
```

```python
#load dataset
def load_data():
    data = pd.read_excel('reviews.xlsx')
    pd.set_option('display.max_colwidth', None)
    return data
sentence_df = load_data()
data = sentence_df
data
```

```python
#definisi dataframe
df = pd.DataFrame(data)
df
```

```python
def remove(sentence):
    sentence = re.sub(r'[0-9]', ' ', sentence)
    sentence = re.sub(r'[^\w\s]', ' ', sentence)
    sentence = re.sub(r'[^A-Za-z\s]', '', sentence)
    sentence = sentence.lower()
    sentence = sentence.strip()
    sentence = re.sub(r'\s+', ' ', sentence)
    sentence = sentence.replace('\n', '')
    sentence = sentence.replace('_', '')
 
    return sentence
df['cleaning'] = df['Reviews'].apply(lambda x: remove(x))
df
```

```python
from nltk.tokenize import word_tokenize
def tokenize(sentence):
    return word_tokenize(sentence)

df_proces = df['cleaning']
df_proces = [tokenize(sentence) for sentence in df_proces]
df['tokenization'] = df['cleaning'].apply(lambda x: tokenize(x))
df['tokenization'] = df_proces
tokens = df_proces
df
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords


stopwords_dictionary = [
    'gak',
    'masa',
    'bisa',
    'lagi',
    'banget',
    'sama',
    'nya',
    'saya',
    'semua',
    'kalo',
    'saat',
    'sambil',
    'ya',
    'untuk',
    'segitu',
    'lain',
    'sih',
    'sangat',
    'tidak',
    'yang',
    'tapi',
    'itu',
    'aduh',
    'lah',
    'buat',
    'mah',
    'tahu',
    'apa',
    'mau',
    'banyak',
    'di',
    'karena',
    'bakal',
    'padahal',
    'ni',
    'orang',
    'terus',
    'lain',
    'sini',
    'hanya',
    'dengan',
    'aja',
    'dan',
    'ada',
    'sekali',
    'udh',
    'kali',
    'walaupun',
    'pdhl',
    'dari',
    'cuma',
    'juga',
    'sesuai',
    'ini',
    'jadi',
    'tt',

]

def remove_stopwords(tokens):
    factory = StopWordRemoverFactory()
    stopwords_remover = factory.create_stop_word_remover()
    combined_stopwords = set(stopwords.words('indonesian')) | set(stopwords_dictionary)

    filtered_tokens = [token for token in tokens if token.lower() not in combined_stopwords]
    return filtered_tokens

df['stopwords'] = df['tokenization'].apply(lambda x: remove_stopwords(x))
df
```

```python
def stemming(tokens):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()

    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return stemmed_tokens

df['stemming'] = df['stopwords'].apply(lambda x: stemming(x))

df
```

```python
import sklearn
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Contoh data
data = df['stemming'].apply(lambda tokens: ' '.join(tokens))

# Membuat objek CountVectorizer
vectorizer = CountVectorizer()

# Mengubah data menjadi term frequency matrix
tf_matrix = vectorizer.fit_transform(data)

# Mendapatkan daftar kata-kata unik (fitur)
features = vectorizer.get_feature_names_out()

# Membuat dataframe dari term frequency matrix
df_tf = pd.DataFrame(tf_matrix.toarray(), columns=features)

# Menampilkan dataframe
print(df_tf)
```

```python
from sklearn.feature_selection import SelectKBest, chi2

# Matriks term frequency sebelumnya
tf_matrix = df_tf

# Label kategori/class dari data
labels = df['Label']

# Hitung skor chi-square
chi2_scores = chi2(tf_matrix, labels)[0]

# Tentukan jumlah fitur teratas yang ingin dipilih
total_features = tf_matrix.shape[1]
k = int(0.5 * total_features)

# Pilih k fitur teratas berdasarkan skor chi-square
selector = SelectKBest(chi2, k=k)
selected_matrix = selector.fit_transform(tf_matrix, labels)

# Tampilkan indeks fitur terpilih
selected_features_indices = selector.get_support(indices=True)
print("Indeks fitur terpilih:")
print(selected_features_indices)

#Dapatkan daftar fitur terpilih
selected_features = [features[i] for i in selected_features_indices]

# Buat dataframe baru dengan fitur terpilih
df_selected_features = pd.DataFrame(selected_matrix, columns=selected_features)

# Simpan hasil seleksi fitur ke dalam dataframe
df_selected_features['Label'] = labels

# Menampilkan dataframe hasil seleksi fitur
print(df_selected_features)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split

# Memisahkan fitur (X) dan label (y)
X = df_selected_features           # Fitur hasil ekstraksi TF
y = df['Label']   # Label kategori/class dari data
  

# Mengubah nilai rating menjadi label kategori/class
y = y.apply(lambda x: 'negatif' if x == 0 else 'positif')

# Membagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)

# Membuat objek model Multinomial Naive Bayes
model = MultinomialNB()

# Melatih model dengan data latih
model.fit(X_train, y_train)

# Menerapkan model pada data uji
y_pred = model.predict(X_test)
print(y_pred)
```

```python
from sklearn import metrics

# Hitung metrik evaluasi
accuracy = metrics.accuracy_score(y_test, y_pred)
precision = metrics.precision_score(y_test, y_pred, average='weighted')
recall = metrics.recall_score(y_test, y_pred, average='weighted')
f1_score = metrics.f1_score(y_test, y_pred, average='weighted')

# Tampilkan hasil
print("Akurasi:", accuracy)
print("Presisi:", precision)
print("Recall:", recall)
print("F1-Score:", f1_score)
```
